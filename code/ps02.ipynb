```python
cos_signal = thinkdsp.CosSignal(440)
duration = cos_signal.period*20.5
cos_wave = cos_signal.make_wave(duration)
cos_wave.plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)
```

```python
cos_spectrum = cos_wave.make_spectrum()
cos_spectrum.plot(high = 50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Hamming Window on the Cosine Signal

```python
cos_wave.hamming()
cos_hamming_spectrum = cos_wave.make_spectrum()
cos_hamming_spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Bartlett Window on the Cosine Signal

```python
cos_wave2 = cos_signal.make_wave(duration)
cos_wave2.window(numpy.bartlett(len(cos_wave2.ys)))
cos_wave2.make_spectrum().plot()
```

Blackman Window on the Cosine Signal

```python
cos_wave3 = cos_signal.make_wave(duration)
cos_wave3.window(numpy.blackman(len(cos_wave3.ys)))
cos_wave3.make_spectrum().plot()
```

Hanning Window on the Cosine Signal

```python
cos_wave4 = cos_signal.make_wave(duration)
cos_wave4.window(numpy.hanning(len(cos_wave4.ys)))
cos_wave4.make_spectrum().plot()
```

Kaiser Window on the Cosine Signal

```python
cos_wave5 = cos_signal.make_wave(duration)
cos_wave5.window(numpy.kaiser(len(cos_wave5.ys),0))
cos_wave5.make_spectrum().plot()
```
