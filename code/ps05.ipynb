### Question 1

#### Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.  The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

The Discrete Fourier Transform Function:

```python
i = complex(0,1)
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

The Inverse Discrete Fourier Transform Function:

```python
def idft(amps):
    N = len(amps)
    ys = dft(amps) / N
    return ys
```

Using Inverse Discrete Fourier Transform to obtain the ys of the wave and plot it:

```python
spectrum_triangle = thinkdsp.TriangleSignal(1).make_wave(duration = 2, framerate=5000).make_spectrum()
ys = idft(numpy.absolute(spectrum_triangle.hs))
thinkdsp.Wave(ys, framerate=11025).plot()
```

New Discrete Fourier Transform that takes frequencies that go from N to 2N-1:

```python
def dft_new(ys):
    N = len(ys) 
    ts = numpy.arange(N,2*N) / N
    freqs = numpy.arange(N,2*N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps

def idft_new(amps):
    N = len(amps)
    ys = dft_new(amps) / N
    return ys
```

Using new Inverse Discrete Fourier Transform to obtain the ys of the wave and plot it:

```python
spectrum_triangle2 = thinkdsp.TriangleSignal(1).make_wave(duration = 2, framerate=5000).make_spectrum()
ys_new = idft_new(numpy.absolute(spectrum_triangle2.hs))
thinkdsp.Wave(ys_new, framerate=11025).plot()
```

There are no differences in using frequencies 0 through N-1, compared to N through 2N-1, becaus we are changing frequencies, not phase of the wave.

### Question 2

#### In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).  Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filter

Functions for creating low-pass and high-pass filters that are each a list of ones accompanied by zeros, or, in the case of the high-pass, a list of zeros accompanied by ones:

```python
def create_filter_low(n,m):
    #n is the number of ones
    #m is the number of zeros
    list1 = n*[1]
    list2 = m*[0]
    list1.extend(list2)
    return list1

def create_filter_high(n,m):
    #m is the number of zeros
    #n is the number of ones
    list1 = n*[0]
    list2 = m*[1]
    list1.extend(list2)
    return list1
```

A low pass filter is made with the create_filter_low function that is then passed into synthesize2, which will take the low_pass_filter as amplitudes and outputs the ys of a window:

```python
low_pass_filter = create_filter_low(10,40)

i = complex(0,1)
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

amps = low_pass_filter
freqs = numpy.linspace(0,2,len(amps))
ts = numpy.linspace(0,10,len(amps)) #time range

ys_syn2 = synthesize2(low_pass_filter,freqs,ts)
thinkplot.plot(ts,ys_syn2)
```

A high pass filter is made with the create_filter_high function that is then passed into synthesize2, which will take the high_pass_filter as amplitudes and outputs the ys of a window:

```python
high_pass_filter = create_filter_high(10,40)

i = complex(0,1)
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

amps_high = high_pass_filter
freqs = numpy.linspace(0,2,len(amps_high))
ts = numpy.linspace(0,10,len(amps_high))

ys_syn2 = synthesize2(amps_high,freqs,ts)
thinkplot.plot(ts,ys_syn2)
```

Convolution corresponds to element-wise multiplication in the frequency domain, which removes the short-term variations from a signal in order to show long-term trends. Applying a smoothing algrithm with a moving average computes the mean of all of the previous n values, for some value of n.

Low-pass filter is an array of ones and then zeros, which drops to zero to eliminate higher frequencies, that only allows lower frequencies to pass through. All low frequencies keep their amplitude and all high frequencies are multiplied by 0 to create a window that is similar to a sinc function.

High-pass filter is an array of zeros and then ones, which starts at zero and then rises to one to allow higher frequencies to pass through. All high frequencies keep their amplitude and all frequencies are byltiplied by 0 to create a window that is similar to a sinc function.

### Question 3

#### So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

Plotting frequencies of a sawtooth signal against the phase offsets:

```python
import scipy.signal
import numpy.fft
sawtooth = thinkdsp.SawtoothSignal(freq=440).make_wave(duration=1, framerate=44100)
ys_sawtooth = sawtooth.ys
segment_sawtooth = sawtooth.segment(duration=0.0035)
amps_sawtooth = numpy.fft.fft(segment_sawtooth.ys)

freqs_original = numpy.arange(len(amps_sawtooth))*(44100/len(ys_sawtooth))
phase_offset_original = numpy.angle(amps_sawtooth)
thinkplot.plot(freqs_original, phase_offset_original)
```

Implementing a gaussian window on the sawtooth signal segment, and then finding the amplitudes to find the phase offsets, which we plot against frequencies:

```python
gaussian_window = scipy.signal.gaussian(M=1,std=0.2)
gaussian_window /= gaussian_window.sum()
smoothed_gaussian = numpy.convolve(segment_sawtooth.ys,gaussian_window, mode='valid')

smoothed_gaussian_amps = dft(smoothed_gaussian)
freqs_new = numpy.arange(len(smoothed_gaussian_amps))*(44100/len(ys_sawtooth))
                                                   
phase_offset_new = numpy.angle(smoothed_gaussian_amps)
thinkplot.plot(freqs_new,phase_offset_new)
```

Implementing a boxcar window on the sawtooth segment, and then finding the amplitudes to find the phase offsets, which we plot against frequencies:

```python
boxcar_window = scipy.signal.boxcar(M=1)
boxcar_window /= boxcar_window.sum()
smoothed_boxcar = numpy.convolve(segment_sawtooth.ys,boxcar_window, mode='valid')

smoothed_boxcar_amps = dft(smoothed_boxcar)
freqs_boxcar = numpy.arange(len(smoothed_boxcar_amps))*(44100/len(ys_sawtooth))
                                                   
phase_offset_boxcar = numpy.angle(smoothed_boxcar_amps)
thinkplot.plot(freqs_boxcar,phase_offset_boxcar)
```

Convolving with windows does not change the phases, because it is shifting and scaling, but not changing the frequency and time values. Increasing the deviation of the windows makes the cutoff frequency decrease, especially for the gaussian window, which changes the phase offset at every mean with different standard deviations, while the boxcar window only varies with mean. The phase offset, however still does not change.

Gaussian window works a lot better, because it has cuttoff frequencies that are higher, and the ratio after the cutoff frequency drops quickly and stays low. (gaussian window looks like a normal distribution plot)

### Question 4

#### Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this window is an increasing function of f, so you can think of diff as a high-pass filter.  One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.  If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.  Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?  See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
def zero_pad(array, n):
    """Makes a new array with the same elements and the given length.

    array: numpy array
    n: length of result

    returns: new NumPy array
    """
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res
```

Computed the DFT of the window [1,-1], and plotted the spectrum:

```python
diff_succ = zero_pad(numpy.asarray([1,-1]),1000)
wave_diff_succ = thinkdsp.Wave(diff_succ, framerate=11025)
wave_diff_succ.make_spectrum().plot()
```

Computed the DFT of the window [-1/2,2,-3/2] and plotted the spectrum:

```python
diff_diff_succ = zero_pad(numpy.asarray([-1/2,2,-3/2]),1000)
wave_diff_diff_succ = thinkdsp.Wave(diff_diff_succ, framerate=11025)
wave_diff_diff_succ.make_spectrum().plot()
```

Computed the DFT of the window [1,-2,1] and plotted the spectrum:

```python
diff_diff_diff_succ = zero_pad(numpy.asarray([1,-2,1]),1000)
wave_diff_diff_diff_succ = thinkdsp.Wave(diff_diff_diff_succ, framerate=11025)
wave_diff_diff_diff_succ.make_spectrum().plot()
```

Computing the DFT of the window [1,-1] is the same as computing the difference between successive elements, and looks and acts as a high pass filter. With each successive order in differentiation, the spectrum looks more and more like a high-pass filter. The second derivative looks the most like high-pass filter, and the order of derivatives follow the backward finite difference pattern.

#### Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

Plotting a parabolic signal spectrum:

```python
parabolic_signal = thinkdsp.ParabolicSignal(freq=440).make_wave()
parabolic_spectrum = parabolic_signal.make_spectrum()
parabolic_spectrum.plot()
```

Plotting the differentiated parabolic spectrum:

```python
differentiated_parabolic_spectrum = parabolic_spectrum
differentiated_parabolic_spectrum.differentiate()
differentiated_parabolic_spectrum.plot()
#parabolic_spectrum.plot()
print(differentiated_parabolic_spectrum)
```

Finding the derivative of a spectrum is the same as multiplying by 2pi(freq)i, which multiplies the amplitude and shifts the phase offset by a certain amount. As shown by the differentiated parabolic spectrum, compared to the original,the differentiated parabolic spectrum is the original, but a lot more amplified (about 2pi(freq) more amplified).
