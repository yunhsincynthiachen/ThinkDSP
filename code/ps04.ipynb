#### 1) Please finish this exercise that we started in class: One of the examples in chap07.ipynb compares two mixtures of complex sinusoids, one with offset=0 and one with offset=1.  The result is a different looking waveform, because "rotating" each component by the same phase offset does not shift each component by the same amount of time.  Can you construct a vector of complex amplitudes that shifts all components by the same amount of time?

```python
#Using matrix multiplication
i = complex(0,1)
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys
```

```python
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = numpy.array([100,200,300,400])

amps2 = amps * numpy.exp(freqs*6.25j)
framerate = 11025
ts = numpy.linspace(0, 1, framerate)

ys = synthesize2(amps, freqs, ts)
ys2 = synthesize2(amps2, freqs, ts)

n=400
thinkplot.plot(ts[:n], ys.real[:n], label=r'$\phi_0 = 0$')
thinkplot.plot(ts[:n], ys2.real[:n], label=r'$\phi_0 = 6.25$')
thinkplot.config(ylim=[-1.15, 1.05])
```

Since we want to construct a vector of complex amplitudes that shifts all components by the same amount of time, we have to shift the entire wave a phase difference. 

In complex sinusoid format, the first wave of no offset is:

cos(2piFT) + isin(2piFT) = e^(i2piFT)

By the properties of sine waves, in order to transform a sine wave, the independent variable (in this case time) must be shifted by an offset:

cos(2piF(T+phi)) + isin(2piF(T+phi)) = e^(i2piF(t+phi))

The exponential form is different than the complex sinusoid of the first wave with no offset:

e^(i2piF(t+phi)) = (e^(i2piFT))X(e^(i2piFphi))

To the right of the multiplication sign (X) is the change in amplitude that needs to be applied to the original wave with no offset.

Synthesize2 provides the matrix multiplication method of the complex sinusoid. This needs to be altered by multiplying the additional (e^i2piFphi). If we were just trying to have a phase offset that rotates the original wave, we would have (e^iphi), but since we are trying to shift the entire wave, we have to consider frequency in all of this. As a result the amplitude of the original wave needs to include frequency in the change.

#### 2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation:  From http://en.wikipedia.org/wiki/Discrete_Fourier_transform:  Read the Wikipedia page on DFT. The discrete fourier transform converts a finite list of equally spaced samples of a function into list of coefficients of finite combination of complex sinusoids, ordered by their frequencies.

```python
i = complex(0,1)
# My implementation of DFT:
def dft_cc(ys):
    total_M = 0
    for n in range(len(ys)):
        args = (1/len(ys))*numpy.arange(len(ys))*n            
        M = ys[n]*numpy.exp(-i * PI2 * args)
        total_M += M #coefficients of finite combination of complex sinusoids
    return total_M

amps = dft_cc(thinkdsp.SawtoothSignal(440).make_wave().ys)
thinkdsp.Spectrum(numpy.absolute(amps), framerate=21000).plot()
thinkplot.show(xlim = [0,5250])
```

#### 3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.  Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

Information about the Cooley-Tukey FFT algorithm:

The length of the list of ys has to be a power of 2. Below, fft_sum_function does a similar job as synthesize2, which synthesizes the Discrete Fourier Transform using matrix multiplication. The fft_ct function splits the sum that was originally in the DFT implementation of number 2 into two lists that are continuously split based off of odd or even frequencies within the length of the list (this is why it needs to be a list with power of 2). 

```python
def fft_sum_function(n,k):
    final = numpy.exp(-PI2*1j*k/n)
    return final
    
def fft_ct(ys):
    N = len(ys)
    if N == 1:
        return ys
    else:
        fft_even = fft_ct(ys[::2])
        fft_odd = fft_ct(ys[1::2])
        all_amps = [0]*N
        for n in xrange(int(N/2)):
            all_amps[n] = fft_even[n] + fft_sum_function(N,n)*fft_odd[n]
            all_amps[n+int(N/2)] = fft_even[n] - fft_sum_function(N,n)*fft_odd[n]
        return all_amps
    
new_amps = fft_ct(thinkdsp.SawtoothSignal(440).make_wave().ys[:2**13])
thinkdsp.Spectrum(new_amps, framerate=22000).plot()
thinkplot.show(xlim = [0,5250])
```

The Cooley-Tukey algorithm in re-expressing the DFT recursively. In order to reduce the computation time of DFT's O(n^2) to O(N log N). This is done by breaking the DFT into smaller DFTs that are performed. The runtime of the cooley-tukey version of FFT runs a lot faster than the DFT in number 3. #### 4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

Allen's implementation of autocorr: 

```python
def serial_corr(wave, lag =1):
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = numpy.corrcoef(y1, y2)[0,1]
    return corr

def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave,lag) for lag in lags]
    return lags, corrs
```

Allen's implementation of fft_autocorr from Chapter 8:

```python
def zero_pad(array, n):
    """Makes a new array with the same elements and the given length.

    array: numpy array
    n: length of result

    returns: new NumPy array
    """
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

Numpy implementation of correlation, reformated to perform autocorrelation below:

```python
def numpy_autocorr(x):
    result = numpy.correlate(x, x, mode='full') #necessary to have correlation with itself
    return result[result.size/2:]
```

Creating the number of loops that will be performed (ns), as well as the signal that will be inputed:

```python
ns = 2 ** numpy.arange(4, 12)
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
print(ns)
```

Running loops of all of the different implementations that I am comparing:

```python
import convolution
results_autocorr = []
results_fft_autocorr = []
results_numpy_autocorr = []
for N in ns:
    ys = noise.ys[:N]
    final_wave = thinkdsp.Wave(ys, framerate=11025)
    result_autocorr = %timeit -o autocorr(final_wave)
    results_autocorr.append(result_autocorr)
    result_fft_autocorr = %timeit -o fft_autocorr(final_wave.ys)
    results_fft_autocorr.append(result_fft_autocorr)
    result_numpy_autocorr = %timeit -o numpy_autocorr(final_wave.ys)
    results_numpy_autocorr.append(result_numpy_autocorr)
```

```python
bests_autocorr = [result.best for result in results_autocorr]
bests_autocorr
bests_fft_autocorr = [result.best for result in results_fft_autocorr]
bests_fft_autocorr
bests_numpy_autocorr = [result.best for result in results_numpy_autocorr]
bests_numpy_autocorr
```

```python
slope_autocorr = scipy.stats.linregress(numpy.log(ns), numpy.log(bests_autocorr))
slope_fft_autocorr = scipy.stats.linregress(numpy.log(ns), numpy.log(bests_fft_autocorr))
slope_numpy_autocorr = scipy.stats.linregress(numpy.log(ns), numpy.log(bests_numpy_autocorr))
[slope_autocorr, slope_fft_autocorr, slope_numpy_autocorr]
```
