#### 2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def serial_corr(wave, lag =1):
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = numpy.corrcoef(y1, y2)[0,1]
    return corr
def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave,lag) for lag in lags]
    return lags, corrs

lags, corr = autocorr(wave)
thinkplot.Plot(lags,corr)
thinkplot.Show(xlabel = 'lags',
               ylabel = 'corr')
```

The autocorrelation function decays exponentially, but then spikes and has a peak after about a lag of 450, so the autocorrelation plot looks like there is a hint of a periodic signal.

#### 4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.  You will find code in ps03starter.ipynb that will help with this.

The functions for analyze1 and analyze2:

```python
def analyze1(ys, freqs, ts):
    args = numpy.outer(ts,freqs)
    M = numpy.cos(thinkdsp.PI2*args)
    amps = numpy.linalg.solve(M,ys)
    return amps
def analyze2(ys,freqs,ts):
    args = numpy.outer(ts,freqs)
    M = numpy.cos(thinkdsp.PI2*args)
    amps = numpy.dot(M,ys)/2
    return amps
```

Running the Algorithmic Complexity of Analyze1:

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

#Arranges equally spaced numbers from 4 to 12
ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)
```

```python
bests = [result.best for result in results]
bests
```

Plot of Analyze1's algorithmic complexity over the equally spaced n-values:

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
x = numpy.log(ns)
y = numpy.log(bests)
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x,y)
[slope, intercept, r_value, p_value, std_err]
```

From the plot above, the slope of the speed of Analyze1 is about 2.076. The calculation of the slope is also shown above. In addition, through the application of least-squares regression, we know that the correlation coefficient is 0.99, which denotes positive correlation. As a result, Analyze1 has a linear characteristic with a slope of about 2.076.

#### 5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this: Break a long signal into segments. Compute the DCT of each segment. Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain. To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT. Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
filename = '100475__iluppai__saxophone-weep.wav'
wave = thinkdsp.read_wave(filename)
wave.make_audio()
```

```python
def remove(elem):
    if abs(elem) < 1000:
        return 0
    else:
        return elem
    
final = wave.segment(0,0)
for i in range(0,10,1):
    seg_small = wave.segment(i,1.0)
    spec_small = seg_small.make_spectrum()
    spec_small.hs = map(remove, spec_small.hs)
    final = final | spec_small.make_wave()
```
