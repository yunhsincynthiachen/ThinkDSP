## 6) A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops to -1 and repeats. See Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.   Note: there is an implementation of SawtoothSignal in thinkdsp.py; you should avoid looking at it until you have written your own.

```python
import math
import numpy
class SawtoothSignal2(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        cycles = self.freq*ts + self.offset / (thinkdsp.PI2)
        frac, _ = numpy.modf(cycles)
        ys = numpy.abs(frac)
        ys = thinkdsp.normalize(thinkdsp.unbias(ys),self.amp)
        return ys
```

```python
SawtoothSignal2(440).make_wave(0.011).plot()
```
